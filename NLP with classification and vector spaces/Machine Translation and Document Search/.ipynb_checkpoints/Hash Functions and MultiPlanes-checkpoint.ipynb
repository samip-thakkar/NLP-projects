# Hash functions and Multi-plane

```python
#BASIC HASH TABLES

#Import the libraries
import numpy as np
import pprint
from utils_nb import plot_vectors
import matplotlib.pyplot as plt

pp = pprint.PrettyPrinter(indent = 4)
```

```python
#Define the basic hash function that give hash values is remainder after division of value and no of buckets

def basic_hash_table(value_list, n_buckets):
    
    #Define a hash function
    def hash_function(value, n_buckets):
        return value % n_buckets
    
    #Create an empty hash table
    hash_table = {i:[] for i in range(n_buckets)}
    
    #Iterate over all values in value list and append in hash table
    for value in value_list:
        #Get the hash value and append in the bucket
        hash_value = hash_function(value, n_buckets)
        hash_table[hash_value].append(value)
    
    return hash_table
```

```python
#Test the function
value_list = [100, 10, 14, 17, 97]
hash_table_example = basic_hash_table(value_list, 10)
pp.pprint(hash_table_example)
```

```python
#Get the scalar sign of the dot product
def side_of_plane(P, v):
    prod = np.dot(P, v.T)
    sign = np.sign(prod)
    scalar_sign = sign.item()
    return scalar_sign
```

```python
#MULTIPLE PLANES

P1 = np.array([[1, 1]])   # First plane 2D
P2 = np.array([[-1, 1]])  # Second plane 2D
P3 = np.array([[-1, 1]]) # Third plane 2D
P_l = [P1, P2, P3]  # List of arrays. It is the multi plane

# Vector to search
v = np.array([[2, 2]])
```

```python
#Define the function that gives final hash value based on the signs of all planes
def hash_multi_plane(P_l, v):
    hash_value = 0
    
    #Iterate through all planes
    for i, P in enumerate(P_l):
        sign = side_of_plane(P, v)
        if sign >= 0:
            hash_i = 1
        else:
            hash_i = 0
        hash_value += 2 ** i * hash_i
    
    return hash_value
```

```python
#Get the hash bucket for vector [2,2]
hash_multi_plane(P_l, v) 
```

```python
#RANDOM PLANES
np.random.seed(0)
num_dimensions = 2 # is 300 in assignment
num_planes = 3 # is 10 in assignment
random_planes_matrix = np.random.normal(
                       size=(num_planes,
                             num_dimensions))
print(random_planes_matrix)
```

```python
v = np.array([[2, 2]])
```

```python
# Side of the plane function. The result is a matrix
def side_of_plane_matrix(P, v):
    dotproduct = np.dot(P, v.T)
    sign_of_dot_product = np.sign(dotproduct) # Get a boolean value telling if the value in the cell is positive or negative
    return sign_of_dot_product
```

```python
sides_l = side_of_plane_matrix(
            random_planes_matrix, v)
sides_l
```

```python
def hash_multi_plane_matrix(P, v, num_planes):
    sides_matrix = side_of_plane_matrix(P, v) # Get the side of planes for P and v
    hash_value = 0
    for i in range(num_planes):
        sign = sides_matrix[i].item() # Get the value inside the matrix cell
        hash_i = 1 if sign >=0 else 0
        hash_value += 2**i * hash_i # sum 2^i * hash_i
        
    return hash_value
```

```python
hash_multi_plane_matrix(random_planes_matrix, v, num_planes)
```
