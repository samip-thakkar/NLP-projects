# Building and Visualizing word frequences

```python
#Import the libraries
import nltk
from nltk.corpus import twitter_samples
import matplotlib.pyplot as plt
import numpy as np
```

```python
nltk.download('stopwords')
from utils import process_tweet, build_freqs
```

```python
#Load sample NLTK data
positive_tweets = twitter_samples.strings('positive_tweets.json')
negative_tweets = twitter_samples.strings('negative_tweets.json')

#Concate both the lists
tweets = positive_tweets + negative_tweets

print(len(tweets))
```

```python
#Make an array of labels
labels = np.append(np.ones((len(positive_tweets))), np.zeros((len(negative_tweets))))
len(labels)
```

```python
#Build a word frequency dictionary
def build_freqs(tweets, ys):
    ylist = np.squeeze(ys).tolist()
    
    freqs = {}
    for y, tweet in zip(ylist, tweets):
        for word in process_tweet(tweet):
            pair = (word, y)
            if pair in freqs:
                freqs[pair] += 1
            else:
                freqs[pair] = 1
    return freqs
```

```python
#Build frequency dictionary
freqs = build_freqs(tweets, labels)
print(len(freqs))
```

```python
#Table of words counts
keys = ['happi', 'merri', 'nice', 'good', 'bad', 'sad', 'mad', 'best', 'pretti',
        '‚ù§', ':)', ':(', 'üòí', 'üò¨', 'üòÑ', 'üòç', '‚ôõ',
        'song', 'idea', 'power', 'play', 'magnific']

data = []

for word in keys:
    pos = 0
    neg = 0
    
    if (word, 1) in freqs:
        pos = freqs[(word, 1)]
    if (word, 0) in freqs:
        neg = freqs[(word, 0)]
        
    data.append([word, pos, neg])
print(data)
```

```python
#Plot the table
fig, ax = plt.subplots(figsize = (10, 10))

#Convert positive raw counts to log + 1(avoid log 1)
x = np.log([x[1] + 1 for x in data])

#Convert negative raw counts to log + 1(avoid log 1)
y = np.log([x[2] + 1 for x in data])

ax.scatter(x, y)

plt.xlabel('Log Positive labels')
plt.ylabel('Log Negative labels')

for i in range(0, len(data)):
    ax.annotate(data[i][0], (x[i], y[i]), fontsize = 12)

ax.plot([0, 9], [0, 9], color = 'red')
plt.show()
```
