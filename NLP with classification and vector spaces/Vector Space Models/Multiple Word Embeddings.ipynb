# Multiple word Embeddings: Find analogies between words

```python
#Import the libraries and files
import numpy as np
import pandas as pd
import pickle

word_embeddings = pickle.load(open('word_embeddings_subset.p', 'rb'))
len(word_embeddings)
```

```python
word_embeddings
```

```python
#We have country vector and corresponding capital vector
countryVector = word_embeddings['country']
print(type(countryVector))
print(countryVector)
```

```python
#Write a function to get a vector of a word
def wordToVec(word):
    return word_embeddings[word]
```

```python
#Sample plots to understand word embeddings of words
import matplotlib.pyplot as plt

words = ['oil', 'gas', 'happy', 'sad', 'city', 'town', 'village', 'country', 'continent', 'petroleum', 'joyful']

#Get the vector represent array of words
bag2d = np.array([wordToVec(word) for word in words])

fig, ax = plt.subplots(figsize = (10, 10))

col1 = 3 #X-axis
col2 = 2 #Y-axis

#Print arrow for word
for word in bag2d:
    ax.arrow(0, 0, word[col1], word[col2], head_width = 0.005, head_length = 0.005, fc = 'r', ec = 'b', width = 1e-5)
    
#Build a scatter plot
ax.scatter(bag2d[:, col1], bag2d[:, col2])

#Add the word label
for i in range(len(words)):
    ax.annotate(words[i], (bag2d[i, col1], bag2d[i, col2]))

plt.show()

#(Village, town), (oil, petrol) are in same and (happy, sad) are in opposite directions
```

# Word distance (Display vector from similar or opposite words)

```python
#Define words
words = ['sad', 'happy', 'town', 'village']

bag2d = np.array([wordToVec(word) for word in words])

fig, ax = plt.subplots(figsize = (10, 10))

col1 = 3
col2 = 2

#Print the arrow
for word in bag2d:
    ax.arrow(0, 0, word[col1], word[col2], head_width = 0.005, head_length = 0.005, fc = 'r', ec = 'r', width = 1e-5)
    
#Print the vector difference between village and town
village = wordToVec('village')
town = wordToVec('town')
diff = town - village
ax.arrow(village[col1], village[col2], diff[col1], diff[col2], fc = 'b', ec = 'b', width = 1e-5)

#Print the vector difference between happy and sad
happy = wordToVec('happy')
sad = wordToVec('sad')
diff = happy - sad
ax.arrow(sad[col1], sad[col2], diff[col1], diff[col2], fc = 'b', ec = 'b', width = 1e-5)

#Prepare a scatter plot
ax.scatter(bag2d[:, col1], bag2d[:, col2])

#Add the word label
for i, word in enumerate(words):
    ax.annotate(word, (bag2d[i, col1], bag2d[i, col2]))

plt.show()
```

```python
#Getting the norm of a vector
print(np.linalg.norm(wordToVec('town')))
print(np.linalg.norm(wordToVec('sad')))
```

# Predicting the countries and capitals

```python
#We can say diff vector between country and capital is vector(country) - vector(capital)
capital = wordToVec('France') - wordToVec('Paris')

#Country is diff + vector(capital)
#Country of Madrid
country = wordToVec('Madrid') + capital 

#Print the first five countries
print(country[:5])
```

```python
#Compare the difference vector and Spain
diff = country - wordToVec('Spain')
print(diff[:10])
```

```python
#Create dataframe for facilitating the algebric operations
keys = word_embeddings.keys()
data = []
for key in keys:
    data.append(word_embeddings[key])

embedding = pd.DataFrame(data = data, index = keys)
embedding.head(10)
```

```python
#Create a function to find the closest word to vector of dataframe
def find_closest_word(v, k = 1):
    
    #Calculate the vector difference between each vector with v
    diff = embedding.values - v
    
    #Get the norm of each difference vector
    delta = np.sum(diff * diff, axis = 1)
    
    #Find the index of minimum distance to vector v
    idx = np.argmin(delta)
    
    #Return the vector of that index
    return embedding.iloc[idx].name
```

```python
#Get the closest vector to predicted country vector of Madrid
find_closest_word(country)
```

```python
find_closest_word(wordToVec('Italy') - wordToVec('Rome') + wordToVec('Madrid'))
```

```python
print(find_closest_word(wordToVec('Berlin') + capital))
print(find_closest_word(wordToVec('Beijing') + capital))
```

```python
#Sometimes it doesn't work
print(find_closest_word(wordToVec('Lisbon') + capital))
```
